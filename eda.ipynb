```python
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces and tabs
```

```python
# Manually categorize numerical and categorical variables based on their nature

# Categorical variables (even if stored as integers)
categorical_vars = [
    "Marital status", "Application mode", "Course", "Daytime/evening attendance",
    "Previous qualification", "Nacionality", "Mother's qualification", "Father's qualification",
    "Mother's occupation", "Father's occupation", "Displaced", "Educational special needs",
    "Debtor", "Tuition fees up to date", "Gender", "Scholarship holder", "International", "Target"
]

# Numerical variables (continuous or ordinal in nature)
numerical_vars = [
    "Application order", "Previous qualification (grade)", "Admission grade", "Age at enrollment",
    "Curricular units 1st sem (credited)", "Curricular units 1st sem (enrolled)",
    "Curricular units 1st sem (evaluations)", "Curricular units 1st sem (approved)",
    "Curricular units 1st sem (grade)", "Curricular units 1st sem (without evaluations)",
    "Curricular units 2nd sem (credited)", "Curricular units 2nd sem (enrolled)",
    "Curricular units 2nd sem (evaluations)", "Curricular units 2nd sem (approved)",
    "Curricular units 2nd sem (grade)", "Curricular units 2nd sem (without evaluations)",
    "Unemployment rate", "Inflation rate", "GDP"
]

import math

# Combine numerical and categorical variables for plotting
all_vars = numerical_vars + categorical_vars

# Determine number of rows needed for an n x 3 layout
cols = 3  # Number of columns
rows = math.ceil(len(all_vars) / cols)  # Determine the number of rows needed

# Create a single figure with an n x 3 layout
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(18, 5 * rows))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot histograms for numerical variables
for i, col in enumerate(numerical_vars):
    sns.histplot(df[col], kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f"Histogram of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

# Plot bar plots for categorical variables
for j, col in enumerate(categorical_vars, start=len(numerical_vars)):
    sns.countplot(x=df[col], ax=axes[j])
    axes[j].set_title(f"Bar plot of {col}")
    axes[j].set_xlabel(col)
    axes[j].set_ylabel("Count")
    axes[j].tick_params(axis='x', rotation=45)

# Remove any unused subplots (in case of an uneven number of plots)
for k in range(len(all_vars), len(axes)):
    fig.delaxes(axes[k])

# Adjust layout and display the figure
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Combine 'Graduate' and 'Enrolled' into a single class 'Not Dropout'
df["Target"] = df["Target"].replace({"Graduate": "Not Dropout", "Enrolled": "Not Dropout"})

# Convert "Target" and categorical variables to categorical type
df["Target"] = df["Target"].astype(str)
df[categorical_vars] = df[categorical_vars].astype(str)  # Ensure categorical vars are strings

# Selected meaningful numerical variables for bivariate analysis
selected_vars = [
    "Admission grade", "Age at enrollment", "Scholarship holder",
    "Curricular units 1st sem (grade)", "Curricular units 2nd sem (grade)",
    "Previous qualification (grade)", "Unemployment rate", "Inflation rate", "GDP"
]

# Determine the number of rows needed
cols = 3  # Number of columns
total_plots = len(selected_vars) + len(categorical_vars)  # Total subplots needed
rows = math.ceil(total_plots / cols)  # Calculate required rows

# Create figure with an n x 3 layout
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(18, 5 * rows))
axes = axes.flatten()  # Flatten the axes array for iteration

# Plot boxplots for numerical variables
for i, col in enumerate(selected_vars):
    sns.boxplot(x=df["Target"], y=df[col], ax=axes[i])
    axes[i].set_title(f"Boxplot of {col} by Target")
    axes[i].set_xlabel("Target (Dropout vs Not Dropout)")
    axes[i].set_ylabel(col)

# Plot bar plots for categorical variables
for j, col in enumerate(categorical_vars, start=len(selected_vars)):
    sns.countplot(x=df[col], hue=df["Target"], ax=axes[j])
    axes[j].set_title(f"Bar plot of {col} by Target")
    axes[j].set_xlabel(col)
    axes[j].set_ylabel("Count")
    axes[j].legend(title="Target")
    axes[j].tick_params(axis='x', rotation=45)

# Remove any unused subplots
for k in range(total_plots, len(axes)):
    fig.delaxes(axes[k])

# Adjust layout and display the figure
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = df[selected_vars].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
```
