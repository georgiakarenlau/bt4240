# Transformation

```python
categorical_vars = [
    "Marital status", "Application mode", "Course", "Daytime/evening attendance",
    "Previous qualification", "Nacionality", "Mother's qualification",
    "Father's qualification", "Mother's occupation", "Father's occupation",
    "Displaced", "Educational special needs", "Debtor", "Tuition fees up to date",
    "Gender", "Scholarship holder", "International"
]


# Function to apply 'others' grouping for categorical variables
def process_categorical_variable(df, column_name):
    value_counts = df[column_name].value_counts()
    if len(value_counts) > 4:
        top_categories = value_counts.head(4).index
        df[column_name] = df[column_name].apply(lambda x: x if x in top_categories else 'others')
    # If categories are less than or equal to 4, no need to group
    return df

# print("check", df["Marital status"].value_counts())
# df.info()

# Apply the process to all categorical variables
for col in categorical_vars:
    df = process_categorical_variable(df, col)

print("Before encoding:", df.columns)
df_encoded = pd.get_dummies(df, columns=categorical_vars, drop_first=True)
print("After encoding:", df_encoded.columns)

# Display the transformed dataset
print(df_encoded.head())
display(df_encoded)
print("Data transformation and one-hot encoding completed successfully!")

df = df.replace({ 'Target' : { 'Dropout' : 1, 'Not Dropout' : 0} })
df['Target'].head()
```

# Model Training XGBoost

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
import numpy as np
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier
import pandas as pd

# Load datasets
train_data = pd.read_csv("..\\BT4240\\train_df.csv")
test_data = pd.read_csv("..\\BT4240\\test_df.csv")

# Define target column
target_column = "Target"
X_train = train_data.drop(columns=[target_column])
y_train = train_data[target_column]

X_test = test_data.drop(columns=[target_column])
y_test = test_data[target_column]

# Define hyperparameter grid
param_grid = {
    "n_estimators": [100, 300],
    "learning_rate": [0.05, 0.1, 0.3],
    "max_depth": [3, 5, 7],
    "min_child_weight": [1, 3],
    "gamma": [0, 0.1, 0.2],
    "subsample": [0.8, 1],
    "colsample_bytree": [0.8, 1],
    "reg_lambda": [1, 10],
    "reg_alpha": [0, 1],
}

# 5-Fold Cross Validation Setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Scoring dictionary
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='macro'),
    'recall': make_scorer(recall_score, average='macro'),
    'f1': make_scorer(f1_score, average='macro')
}

# Initialize XGBoost model
xgb_model = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
)

# GridSearchCV: Perform hyperparameter tuning
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_grid,
    n_iter=30,  # Only test 30 random combinations instead of 864
    scoring=scoring,
    refit="f1",
    cv=cv,
    n_jobs=-1,
    verbose=1,
    return_train_score=True,
    random_state=42
)

random_search.fit(X_train, y_train)

# Get the results and format them
results_df = pd.DataFrame(random_search.cv_results_)
top_5 = results_df.nlargest(5, "mean_test_f1")

# Show top 5 models with all metrics
metrics = ["mean_test_accuracy", "mean_test_precision", "mean_test_recall", "mean_test_f1"]
top_5_display = top_5[["params"] + metrics]

pd.set_option("display.max_colwidth", None)
print("Top 5 Best Parameter Sets by F1-Score (with all metrics):")
print(top_5_display.to_string(index=False))

best_model = random_search.best_estimator_
best_model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = best_model.predict(X_test)

# Evaluate performance
test_accuracy = accuracy_score(y_test, y_pred)
test_precision = precision_score(y_test, y_pred, pos_label=1)
test_recall = recall_score(y_test, y_pred, pos_label=1)
test_f1 = f1_score(y_test, y_pred, pos_label=1)

print(f'Test Accuracy: {test_accuracy:.4f}, Test Precision: {test_precision:.4f}, '
      f'Test Recall: {test_recall:.4f}, Test F1: {test_f1:.4f}')
```
